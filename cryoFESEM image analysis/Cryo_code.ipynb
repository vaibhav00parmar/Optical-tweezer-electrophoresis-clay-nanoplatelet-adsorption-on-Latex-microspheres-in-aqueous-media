```python
import cv2
%matplotlib notebook
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# resd the image 
img = cv2.imread('sample_fig.tif') 
plt.imshow(img)
```

```python
#---creating blank image of same size---#
blank=np.zeros(img.shape[:2],dtype='uint8')
#plt.imshow(blank)
```

```python
#---creating a white spot on the blank image---#
circle=cv2.circle(blank,(467,400),280,255,-1)  #---((cx,cy),radius,gray_val, thickness)
#plt.imshow(circle)
```

```python
#---masking the created spot on the main image---#
masked=cv2.bitwise_and(img,img,mask=circle)
plt.imshow(masked)
```

```python
# Use the cvtColor() function to grayscale the image
gray_image = cv2.cvtColor(masked, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

```python
C=[]
for i in range (0,len(gray_image),1):
   for j in range (0,len(gray_image[0]),1):
     if(gray_image[i][j]!=0):
         C.append(gray_image[i][j])   
```

```python
c=np.array(C)
print(c)
print("mean of intensity: ", np.mean(c))
print("std of arr: ", np.std(c))
```
